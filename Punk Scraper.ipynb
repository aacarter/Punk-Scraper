```python
import requests
page = requests.get("https://www.larvalabs.com/cryptopunks/forsale")
page
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
links = soup.find_all('a')
titles = [link.get('title') for link in links]
titles
```

```python
import pandas as pd
forsale = pd.DataFrame({
    "Description": titles
})
forsale.head()
```

```python
df = forsale.dropna()
```

```python
texttocolumns = df['Description'].str.split(' ',expand=True)
texttocolumns
```

```python
texttocolumns = texttocolumns.drop(texttocolumns.columns[[0,2,3,5,7,8,9,10,11,12,13]], axis=1)
texttocolumns.head()
```

```python
texttocolumns.columns =['CryptoPunkId', 'ETH', 'Price']
```

```python
texttocolumns['CryptoPunkId'] = texttocolumns['CryptoPunkId'].str.replace('#', '')
texttocolumns['Price'] = texttocolumns['Price'].str.replace('(', '')
texttocolumns['Price'] = texttocolumns['Price'].str.replace('$', '')
texttocolumns['Price'] = texttocolumns['Price'].str.replace(',', '')
texttocolumns.head()
```

```python
texttocolumns.to_csv('punksforsale.csv', index = False)
```
